```
def group_and_barplot(feat_groupby, feat_agg, agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    #.agg(np.mean)
    #.agg(np.median)
    .agg(agg_funcs)
    .sort_values(by=feat_sort, ascending=False)
    .head(top) #50 most expensive
  ).plot(kind='bar', figsize=(20, 10), subplots=subplots); #; hides matplot output
```

```
group_and_barplot('param_marka-pojazdu', 'price_value');
```

```
group_and_barplot('param_rok-produkcji', 'price_value', feat_sort='median');
```
